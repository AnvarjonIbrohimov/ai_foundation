```python
import numpy as np
```

```python
arr = np.array([1,3,5,2])
print("numpy_array:", arr)
```

```python
arr2 = np.array([
    [1,3,5,2],
     [2,1,4,2]
    ])
print("numpy_array:", arr2)
```

```python
# dtype  qaysi type ekanni chiqarib beradi
a = np.array([1,2,3])
print(a.dtype)
```

```python
# reshape  2 row 3 columnga ozgartirib beradi
a = np.array([1,2,3,4,5,6])
b = a.reshape(2,3)
print("shaped_array:", b)
```

```python
# astype() typeni ozgartirib beradi
a = np.array([1,2,3])
b = a.astype(float)
print("float:", b)
```

```python
# np.zeros() typeni ozgartirib beradi
z = np.zeros((2,3))
o = np.ones((2,3))
print("zeros:", z)
print("ones:", o)
```

```python
# ndin
a = np.array([[1,2], [3,4]])
print("ndin:", a.ndim)
```

```python
# arrange() => ketma ket array (1,6) 1 dan 6 gacha chiqarib ebradi 6 chiqmaydi
a = np.arange(1,6)
print("arange:", a)
```

```python
# sum() mean() max() min()
a = np.array([1,2,3,4,5,6])

print("sum:", a.sum())
print("mean:", a.mean())
print("max:", a.max())
print("min:", a.min())
```

```python
# flatten( )
a = np.array([[1,2], [3,4]])
print('flatten:', a.flatten())
```

```python
# T( )
a = np.array([[1,2], [3,4]])
print('transpose:', a.T)
```

```python
# dor() @
a = np.array([[1,2], [3,4]])
b = np.array([[2,4], [1,2]])
print("dot/multiplication:", a.dot(b))
```

```python
# random.rand()
a = np.random.rand(2,4) # => 0.0 / 1.0 included
print("random.rand(): ", a)
```
