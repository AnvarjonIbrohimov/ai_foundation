```python
data_with_missing = {
    "cars": ["BMW", "Volvo", None, "Ford", "Chevrolet", "Genisis"],
    "counts": [2,3,None,5,2,4]
}
df2 = pd.DataFrame(data_with_missing)
print(df2)
```

```python
df2.isnull().sum() # null qiymat bormi bolsa sum() yani nechta?
```

```python
df2.dropna() # null qiymatdagi qatorlarni ochirib tashlaydi
```

```python
print(df2)
```

```python
df2.fillna("empty") # null qiymatlar orniga shu qiymatni qoyib beradi!
```
